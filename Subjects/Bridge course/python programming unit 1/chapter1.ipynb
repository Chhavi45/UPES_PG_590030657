### Python String Built-in Methods (Full List)

#### lower()

Return a copy of the string with all the cased characters converted to lowercase. For example:

```python
a = "PYTHON"
a.lower()
```

#### upper()

Return a copy of the string with all the cased characters converted to uppercase. Note that s.upper().isupper() might be False if s contains uncased characters or if the Unicode category of the resulting character(s) is not “Lu” (Letter, uppercase), but e.g. “Lt” (Letter, titlecase).

```python
"python".upper()
```

#### title()

Return a title cased version of the string where words start with an uppercase character and the remaining characters are lowercase.



```python
"python programming language".title()
```

#### capitalize()

The capitalize() Return a copy of the string with its first character capitalized and the rest lowercased.

```python
"python programming".capitalize()
```

#### swapcase()


The swapcase() method returns a string where all the upper case letters are lower case and vice versa.

```python
"pYtHOn".swapcase()
```

#### casefold() (stronger lower)


Return a casefolded copy of the string. Casefolded strings may be used for caseless matching.

Casefolding is similar to lowercasing but more aggressive because it is intended to remove all case distinctions in a string. For example, the German lowercase letter 'ß' is equivalent to "ss". Since it is already lowercase, lower() would do nothing to 'ß'; casefold() converts it to "ss". For example:

```python
"GERMAN ß".casefold() 
```

#### strip(chars=None)

Remove Spaces / Cleanup

```python
"    hello    Chhavi".strip()
```

#### lstrip(chars=None)

Return a copy of the string with leading characters removed.


```python
"  hello".lstrip()
```

#### rstrip

Return a copy of the string with trailing characters removed.

```python
"hello    ".rstrip()
```

## 1) Numeric Data Types in Python

Python provides three main numeric data types:

int → Whole numbers (no decimal)
float → Decimal numbers (real numbers)
complex → Complex numbers like a + bj

Additionally, bool is also treated as a numeric type in Python.


```python
a = 10
b = -99
c = 10.3
d = -13.4

print("a =", a, "| type:", type(a))
print("b =", b, "| type:", type(b))
print("c =", c, "| type:", type(c))
print("d =", d, "| type:", type(d))

print("a + b =", a + b)
print("a * 5 =", a * 5)

print("c + d =", c + d)
print("c * 2 =", c * 2)

z = 2 + 3j

print("z =", z, "| type:", type(z))
print("Real part:", z.real)
print("Imaginary part:", z.imag)
```

Constants: UPPER_CASE,
Classes: PascalCase,
Protected (internal use): _single_leading_underscore,
Strongly "private" (name mangling): __double_leading_underscore,
Special (magic/dunder): __init__, __str__

```python
#Variables/functions (snake_case)
student_name = "Chhavi"
total_marks = 95

def calculate_average(a, b):
    return (a + b) / 2

# Constants (UPPER_CASE)
PI = 3.14159
MAX_RETRIES = 3

#Class (PascalCase)
class StudentProfile:
    pass

#Internal use
_internal_cache = {}

#Name mangling (for class memebers)
class BankAccount:
    def __init__(self):
        self._balance = 0  # name-mangled
        
print(student_name, total_marks, calculate_average(10, 20), PI, MAX_RETRIES)        
```

Used to check whether two references point to the same object in memory.

#### 4) Operator Precedence & Associativity

Precedence decides which operator is evaluated first in an expression.

General ideas:

() parentheses have highest precedence

** (power) is high precedence and is right-associative

* / // % come before + -

Comparisons come after arithmetic

not > and > or


When in doubt, use parentheses to make intention clear.

```python
print(2 + 3 * 4)

print((2 + 3) * 4)

print(-3 ** 2)
print((-3) ** 2)

print(2 ** 3 ** 2)
```

#### Logical precedence: not, and, or

Evaluation order:

1. not
2. and
3. or

```python
a = True
b = False
c = False

print(not a and b)

print(a or b and c)
```

### 5) type() and id() in Python

#### type(object)

Returns the data type (class) of a value/object.

#### id(object)

Returns a unique identity number for the object (often its memory address in CPython). Useful for understanding:

object identity, 
mutability, 
references

```python
print(type(10))
print(type(10.5))
print(type("python"))
print(type([1, 2, 3]))
print(type({"a": 1}))
print(type(True))
print(type(None))
```

```python
x = 10
y = 10
z = 11

print("id(x) =", id(x))
print("id(y) =", id(y))
print("id(z) =", id(z))

print("x is y?", x is y)
print("x == y?", x == y)
```

#### id() and Mutability (Very Important)

1. Immutable types: int, float, str, tuple
2. If you "change" them, Python creates a new object.
3. Mutable types: list, dict, set
4. They can be modified in-place without creating a new object.


```python
a = 100
print("Before change:", a, "id:", id(a))

a = a + 1
print("After change:", a, "id:", id(a))
```

```python
lst = [1, 2, 3]
print("Before change:", lst, "id:", id(lst))

lst.append(4)
print("After append:", lst, "id:", id(lst))  # same id
```

### Equality (==) vs Identity (is)

1. == checks whether values are equal
2. is checks whether two variables point to the same object

Use:

1. == for value comparison
2. is for checking None or object identity

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print("a == b:", a == b)  # True (values same)
print("a is b:", a is b)  # False (different objects)
print("a is c:", a is c)  # True (same object)

val = None
print("val is None:", val is None)  # recommended way
```
