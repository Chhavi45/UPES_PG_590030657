### Lecture 1 — Strings
#### 1) Initialization

Strings are immutable sequences of characters.

Create using:

Single quotes: 'hello'

Double quotes: "hello"

Triple quotes (multi-line): """hello"""

```python
s1 = 'Python'
s2 = "Data Science"
s3 = """Line1
Line2"""

print(s1)
print(s2)
print(s3)
```

### 3) Indexing (0-based)


```python
s = "Python"
print(s[0])
print(s[-1])

print(s[len(s)-1])
```

### 4) Slicing
Form: s[start : stop : step] (stop excluded)

```python
s = "Python"
print(s[1:4])
print(s[:3])
print(s[3:])
print(s[5:1:-1])
```

### Lists

#### 1) Initialization
Lists are mutable, ordered collections.

```python
L = [10, 20, 30]
mixed = [1, "A", 3.5, True]
empty = []
print(L)
print(mixed)
print(empty)
```

#### 2) Common methods

append(x), extend(iterable), insert(i,x)
remove(x), pop(i), clear()
index(x), count(x)
sort(), reverse()

```python
L = [3, 1, 2]
L.append(5)         # [3,1,2,5]
print(L)
L.sort()            # [1,2,3,5]
print(L)
L.pop(2)
print(L)
L.remove(5)
print(L)
L.clear()
print(L)
```

#### 3) Operations

```python
print([1,2] + [3,4])
print([0]*4)
print(2 in [1,2,3])
```

#### 4) List comprehension
Compact way to build lists.

```python
squares = [x*x for x in range(1,6)]
evens = [x for x in range(10) if x%2==0]
pairs = [(x,y) for x in [1,2] for y in [3,4,6]]
print(squares)
print(evens)
print(pairs)
```

#### 5) Nesting
List of lists (2D structure).

```python
mat = [[1,2,3],[4,5,6]]
print(mat[1][2])   # 6
```

### Lecture 4 — Sets & Dictionaries
### A) Sets
#### 1) Initialization
Sets are unordered, unique elements

```python
S = {1, 2, 3}
S2 = set([1,2,2,3])
empty_set = set()
print(S)
print(S2)
print(empty_set)
```

#### 2) Methods / Operations
add, remove, discard, pop
union(|), intersection(&), difference(-), symmetric_difference(^)

```python
A = {1,2,3}
B = {3,4,5}
print(A)
print(B)
print(A | B)
print(A & B)
print(A - B)
```

#### B) Dictionaries
##### 1) Initialization
Key–value mapping. Keys must be hashable (immutable types).

```python
d = {"name":"Chhavi", "age":21}
d2 = dict(a=1, b=2)
print(d)
print(d2)
```

#### 2) Common operations/methods
Access: d[key], safe access: d.get(key, default).  
Insert/update: d[key] = value, update(). 
Remove: pop(key), popitem(), del d[key].  
Views: keys(), values(), items()

```python
print(d)
d["age"] = 21
print(d)
print(d.get("city", "NA"))
```

#### 3) Nesting

```python
student = {
    "name":"A",
    "marks":{"math":90, "python":95}
}
student["marks"]["python"]
```

#### 4) Sorting & typecasting
Sort dict by keys or by values using sorted()

```python
a = "JDEJEBANDAMLKS"
sorted(a)
```

```python
d = {"b":2, "a":5, "c":1}
print(sorted(d))
print(sorted(d.items()))
print(sorted(d.items(), key=lambda x: x[1]))
```

#### 5) Recursion
Function calls itself; must have base case.

```python
def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)
```

#### 4) Inner functions + returning functions

```python
def outer(msg):
    def inner():
        return msg.upper()
    return inner

f = outer("hello")
print(f)
print(f())
```

```python
outer("help")()
```

#### 5) Passing collections/functions

```python
def apply_twice(f, x):
    return f(f(x))

apply_twice(lambda a: a+1, 5)
```

### Lecture 7 — Mutability in functions + Applications + Doubts

#### 1) Mutable vs Immutable behavior
Immutable types: int, float, str, tuple Mutable types: list, dict, set

##### Key idea:
If you modify a mutable object inside a function, the caller sees it.
If you “modify” immutable objects, you actually create a new object.

```python
def change_num(x):
    x = x + 1
    
def change_list(L):
    L.append(99)
    
a = 10
change_num(a)
print(a)

L = [1,2]
change_list(L)
print(L)        
```
