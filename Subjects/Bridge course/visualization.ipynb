```python
df["Average_Marks"] = (
    df["Math_Marks"] + df["Python_Marks"]
) / 2
```

```python
df["Average_Marks"] = df["Average_Marks"].round(2)
```

```python
sns.barplot(
    data=df,
    x="Student",
    y="Python_Marks"
)

plt.title("python marks of student")
plt.xlabel("Student Name")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.barplot(
    data=df,
    x="Student",
    y="Python_Marks"
)

plt.title("python Marks of students")
plt.xlabel("Student Name")
plt.ylabel("Python Marks")

plt.xticks(rotation=45)

plt.show()
```

```python
sns.barplot(
    data=df,
    x="Department",
    y="Python_Marks"
)

plt.title("Average python marks by Department")
plt.xlabel("department")
plt.ylabel("Average python Marks")
plt.show()
```

```python
sns.countplot(
    data=df,
    x="Department"
)

plt.title("Number of students in each department")
plt.xlabel("Department")
plt.ylabel("Number of students")
plt.show()
```

```python
sns.histplot(
    data=df,
    x="Python_Marks"
)

plt.title("Distribution of python Marks")
plt.xlabel("Python Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(
    data=df,
    x="Python_Marks",
    kde=True
)

plt.title("Distribution of Python Marks")
plt.xlabel("Python Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(
    data=df,
    x="Python_Marks",
    bins=5,
    kde=True
)

plt.title("Distribution of Python Marks")
plt.show()
```

```python
sns.boxplot(
    data=df,
    y = "Python_Marks"
)

plt.title("Box plot of Python marks")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.boxplot(
    data=df,
    x="Department",
    y="Python_Marks"
)

plt.title("Python Marks by Department")
plt.xlabel("Department")
plt.ylabel("Python_Marks")
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x="Attendance",
    y="Python_Marks"
)

plt.title("Attendance versus Python marks")
plt.xlabel("Attendance")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x="Attendance",
    y="Python_Marks",
    hue="Department"
)

plt.title("Attendance versus python Marks")
plt.xlabel("Attendance")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x="Attendance",
    y="Python_Marks",
    hue="Department",
    size="Average_Marks"
)

plt.title("Attendance, Python Marks and Average Marks")
plt.show()
```
